```python

def skewness(dist):
    """Calculates the centralised and normalised skewness of dist."""
    average = np.mean(dist)
    std_dev = np.std(dist)
    value = np.sum(((dist - average) / std_dev)**3) / len(dist - 2)
    return value

def kurtosis(dist):
    """Calculates the centralised and normalised excess kurtosis of dist."""
    average = np.mean(dist)
    std_dev = np.std(dist)
    value = np.sum(((dist - average) / std_dev)**4) / len(dist - 3) - 3.0
    return value

def bootstrap(dist, function, confidence_level=0.90, nboot=10000):
    """Carries out a bootstrap of dist to get the uncertainty of a statistical function applied to it."""
    fvalues = np.array([])

    dlen = len(dist)
    for _ in range(nboot):
        random_sample = np.random.choice(dist, dlen, replace=True)
        f = function(random_sample)
        fvalues = np.append(fvalues, f)

    # Lower and upper quantiles
    qlow = 0.5 - confidence_level / 2.0
    qhigh = 0.5 + confidence_level / 2.0
    low = np.quantile(fvalues, qlow)
    high = np.quantile(fvalues, qhigh)

    return low, high

# Transpose the years column
def transpose_years(df):
    df_years = df.iloc[:, 4:].transpose()
    df_years.columns = df['Country Name']
    return df_years

# Check whether the module is imported or run directly
if __name__ == "__main__":
    # Example usage with a random normal distribution
    dist = np.random.normal(4.0, 3.0, 10000)

    print("skewness =", np.round(skewness(dist), 6))
    print("kurtosis =", np.round(kurtosis(dist), 6))

    print()
    # Call the bootstrap routine with statistical functions
    low, high = bootstrap(dist, np.mean, confidence_level=0.682)
    sigma = 0.5 * (high - low)
    print("average =", np.round(np.mean(dist), 4), "+/-", np.round(sigma, 4))

    low, high = bootstrap(dist, np.std, confidence_level=0.682)
    sigma = 0.5 * (high - low)
    print("std. dev =", np.round(np.std(dist), 4), "+/-", np.round(sigma, 4))

    low, high = bootstrap(dist, skewness, confidence_level=0.682)
    sigma = 0.5 * (high - low)
    print("skewness =", np.round(skewness(dist), 4), "+/-", np.round(sigma, 4))

    low, high = bootstrap(dist, kurtosis, confidence_level=0.682)
    sigma = 0.5 * (high - low)
    print("kurtosis =", np.round(kurtosis(dist), 4), "+/-", np.round(sigma, 4))

    # Example of transposing years
    # Replace 'your_dataframe.csv' with the actual path to your dataframe
    df = pd.read_csv('your_dataframe.csv')
    df_transposed = transpose_years(df)
    print("\nTransposed Years:")
    print(df_transposed)
```
